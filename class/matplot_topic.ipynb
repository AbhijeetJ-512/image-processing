```python
import matplotlib.pyplot as plt
import random
```

```python
x = [x for x in range(11)]
y = [x*2 for x in range(11)]
```

```python
plt.plot(x,y,color='red')
plt.show()
plt.bar(x,y)
plt.show()
plt.pie(x)
plt.show()
```

```python
plt.figure(figsize=(10,10),facecolor='cyan'     )
plt.subplot(2,2,1)
plt.title('bar graph')
plt.bar(x,y)
plt.subplot(2,2,2)
plt.title('pie graph')
plt.pie(x,autopct='%1.2f%%')
plt.subplot(2,2,3)
plt.title('scatter graph')
plt.scatter(x,y)
plt.subplot(2,2,4)
plt.title('line graph')
plt.plot(x,y)
```
